```python
import numpy as np
import pandas as pd
```

```python
import glob

data_files = glob.glob("./*.csv")

files_list = []

for files in data_files:
    df = pd.read_csv(files, index_col=None, header=0)
    files_list.append(df)
    
census_data = pd.concat(files_list, axis=0, ignore_index=True)
census_data.head(10)
```

```python
census_data.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
census_data.dtypes
```

```python
census_data.head()
```

```python
census_data[['Male', 'Female']] = census_data.GenderPop.str.split("_", expand=True)
```

```python
census_data.head()
```

```python
census_data.drop(columns=["GenderPop","location","product_type","product_description","quantity","price"], inplace=True)
```

```python
census_data.head()
```

```python
census_data["Income"] = census_data.Income.str.strip('$')
```

```python
census_data.head()
```

```python
census_data = census_data.replace('%','',regex=True)
census_data.head()
```

```python
census_data = census_data.replace('M','',regex=True)
census_data.head()
```

```python
census_data = census_data.replace('F','',regex=True)
census_data.head()
```

```python
census_data.dtypes
```

```python
census_data.loc[:,'Hispanic':'Income'] = round(census_data.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
census_data.head(20)
```

```python
census_data.dtypes
```

```python
census_data.info()
```

```python
np.where(census_data.applymap(lambda x: x == ''))
```

```python
census_data['Male'] = census_data['Male'].astype(float)
```

```python
census_data.drop(columns=['Female'],inplace=True)
```

```python
census_data['Female'] = census_data['TotalPop'] - census_data['Male']
```

```python
census_data
```

```python
np.where(census_data.applymap(lambda x: x == ''))
```

```python
census_data.dtypes
```

```python
census_data.drop_duplicates(inplace=True)
```
